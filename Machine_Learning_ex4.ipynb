```python
def sigmoid(z):
    return 1.0/(1.0 + np.exp(-z))
```

```python
def forword_propagate(X,theta1,theta2):
    # 样本数量
    m = X.shape[0]
    
    a1 = np.insert(X,0,values=np.ones(m),axis=1)
    z2 = a1*theta1.T
    a2 = sigmoid(z2)
    a2 = np.insert(a2,0,values=np.ones(m),axis=1)
    z3 = a2*theta2.T
    h = sigmoid(z3)
    return a1,z2,a2,z3,h
```

```python
# 实现神经网络的代价函数
def cost(theta1,theta2,X,y,input_size,hidden_size,num_labels,learning_rate):
    # 虽然matrix和ndarray之间也可以进行运算，但是最好还是转化为同一类型，防止混淆
    m = X.shape[0]
    X = np.matrix(X)
    y = np.matrix(y)
    theta1 = np.matrix(theta1)
    theta2 = np.matrix(theta2)
    
    a1,z2,a2,z3,h = forword_propagate(X,theta1,theta2)
    first = -np.multiply(y,np.log(h))
    second = np.multiply((y-1),np.log(1-h))
    J = np.sum(first+second)/m
    return J
```

```python
# sigmoid梯度，即该函数的导数
def sigmoid_gradient(z):
    return np.multiply(sigmoid(z),(1-sigmoid(z)))
```

```python
# 反向传播算法，先求的是不带正则化的反向传播


# 重点！！！！难点！！！！！重点！！！！难点！！！！！重点！！！！难点！！！！！重点！！！！难点！！！！！


def backprop(params,input_size,hidden_size,num_labels,X,y,learning_rate):
    m = X.shape[0]
    X = np.matrix(X)
    y = np.matrix(y)
    
    theta1 = np.array(params[0:(input_size+1)*hidden_size]).reshape((hidden_size,input_size+1))
    theta2 = np.array(params[(input_size+1)*hidden_size:]).reshape((num_labels,hidden_size+1))
    
    # 根据该theta计算出输出值，即先使用前向传播
    a1,z2,a2,z3,h = forword_propagate(X,theta1,theta2)
    
    # 求出代价函数的值
    J = 0
    J = cost(theta1,theta2,X,y,input_size,hidden_size,num_labels,learning_rate)
    
    # 初始化delta为0，为后续累加做准备
    delta1 = np.zeros(theta1.shape)           # 25*401
    delta2 = np.zeros(theta2.shape)           # 10*26
    
    # 进行for循环遍历每一个样本，
    for t in range(m):
        a1t = a1[t,:]        #(1,401)
        z2t = z2[t,:]        #(1,25)
        a2t = a2[t,:]        #(1,26)
        ht = h[t,:]          #(1,10)
        yt = y[t,:]          #(1,10)
        
        d3t = ht-yt          #(1,10)
        
        z2t = np.insert(z2t,0,values=np.ones(1))       # 可以不指定,axis=1
        d2t = np.multiply(d3t*theta2,sigmoid_gradient(z2t))
        
        delta1 += d2t[:,1:].T*a1t
        delta2 += d3t.T*a2t
        
    delta1/=m
    delta2/=m
    
    return J,delta1,delta2    
```

```python
# 正则化神经网络
def backpropReg(params,input_size,hidden_size,num_labels,X,y,learning_rate):
    m = X.shape[0]
    X = np.matrix(X)
    y = np.matrix(y)
    
    theta1 = np.array(params[0:(input_size+1)*hidden_size]).reshape((hidden_size,input_size+1))
    theta2 = np.array(params[(input_size+1)*hidden_size:]).reshape((num_labels,hidden_size+1))
    
    # 根据该theta计算出输出值，即先使用前向传播
    a1,z2,a2,z3,h = forword_propagate(X,theta1,theta2)
    
    # 求出代价函数的值
    J = 0
    J = cost(theta1,theta2,X,y,input_size,hidden_size,num_labels,learning_rate)
    
    # 初始化delta为0，为后续累加做准备
    delta1 = np.zeros(theta1.shape)           # 25*401
    delta2 = np.zeros(theta2.shape)           # 10*26
    
    # 进行for循环遍历每一个样本，
    for t in range(m):
        a1t = a1[t,:]        #(1,401)
        z2t = z2[t,:]        #(1,25)
        a2t = a2[t,:]        #(1,26)
        ht = h[t,:]          #(1,10)
        yt = y[t,:]          #(1,10)
        
        d3t = ht-yt          #(1,10)
        
        z2t = np.insert(z2t,0,values=np.ones(1))       # 可以不指定,axis=1
        d2t = np.multiply(d3t*theta2,sigmoid_gradient(z2t))
        
        delta1 += d2t[:,1:].T*a1t
        delta2 += d3t.T*a2t
        
    delta1/=m
    delta2/=m
    
    delta1[:,1:] = delta1[:,1:]+learning_rate*theta1[:,1:]/m
    delta2[:,1:] = delta2[:,1:]+learning_rate*theta2[:,1:]/m
    grad = np.concatenate((np.ravel(delta1),np.ravel(delta2)))
    
    return J,grad
```

```python
# 实用工具库计算参数的最优解
# 由于每次使用的初始化theta值不相同，故每次训练出来的结果是不同的
fmin = minimize(fun=backpropReg,x0=(params),args=(input_size,hidden_size,num_labels,X,y_onehot,learning_rate),method='TNC',jac=True,options={'maxiter': 250})
fmin
```
