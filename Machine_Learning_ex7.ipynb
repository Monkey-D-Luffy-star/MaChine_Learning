```python
import numpy as np
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
from scipy.io import loadmat
import seaborn as sb         # 这个包适用于绘制更漂亮的图像
```

```python
def find_closed_centroids(X,centroids):
    m = X.shape[0]
    k = centroids.shape[0]
    idx = np.zeros(m)
    
    # 遍历数据集，寻找最接近的centroid
    for i in range(m):
        min_dist = 10000
        for j in range(k):
            dist = np.sum((X[i,:]-centroids[j,:])**2)
            if dist < min_dist:
                idx[i] = j
                min_dist = dist
                pass
            pass
        pass
    return idx
```

```python
# 定义一个K-Means算法，多次随机初始点
def run_k_means(X,centroids_init,max_iters):
    m,n = X.shape
    k = centroids_init.shape[0]
    centroids = centroids_init
    idx = np.zeros(m)
    
    for i in range(max_iters):
        idx = find_closed_centroids(X,centroids)
        centroids = compute_centroids(X,idx,k)
        pass
    return centroids,idx
```

```python
centroids,idx = run_k_means(X,centroids_init,max_iters = 50)
centroids,idx
```

```python
# 下面写出随机初始化均值点的函数，避免手动初始化均值点造成的局部最优
def init_centroids(X,k):
    m,n = X.shape
    centroids = np.zeros((k,n))
    idx = np.random.randint(1,m,k)
    
    for i in range(k):
        centroids[i,:] = X[idx[i],:]
        pass
    return centroids
```

```python
centroids,idx = run_k_means(X,init_centroids(X,k=3),max_iters = 50)
cluster1 = X[np.where(idx==0)[0],:]
cluster2 = X[np.where(idx==1)[0],:]
cluster3 = X[np.where(idx==2)[0],:]
fig,ax = plt.subplots(figsize=(12,8))
ax.scatter(cluster1[:,0],cluster1[:,1],s=30, color='red', label='Cluster 1')
ax.scatter(cluster2[:,0],cluster2[:,1],s=30, color='green', label='Cluster 2')
ax.scatter(cluster3[:,0],cluster3[:,1],s=30, color='blue', label='Cluster 3')
ax.legend(loc='best')
plt.show()
```

```python
# PCA 降维算法
path = 'D:/Google_Download/andrew_ml_ex78376/ex7data1.mat'
data = loadmat(path)
X = data['X']
X[:3,:]
```

```python
# 绘制这些点
fig,ax = plt.subplots(figsize=(12,8))
ax.scatter(X[:,0],X[:,1])
plt.show()
```

```python
def pca(X):
    # 归一化数据
    X = (X-X.mean())/X.std()
    
    # 计算convariance 矩阵
    X = np.matrix(X)
    cov = (X.T*X)/X.shape[0]
    
    U,S,V = np.linalg.svd(cov)
    
    return U,S,V
```

```python
U,S,V = pca(X)
U,S,V
```

```python
def project_data(U,X,k):
    U = np.matrix(U)
    X = np.matrix(X)
    Ureduce = U[:,:k]
    return np.array(Ureduce.T*X.T)
```

```python
Z = project_data(U,X,1)
Z
```

```python
def recover_data(U,Z,k):
    U = np.matrix(U)
    Z = np.matrix(Z)
    Ureduce = U[:,:k]
    return np.array((Ureduce*Z).T)
```

```python
X_recovered = recover_data(U,Z, 1)
X_recovered
```

```python
# 描出这些点
fig,ax = plt.subplots(figsize=(12,8))
ax.scatter(X_recovered[:,0],X_recovered[:,1])
plt.show()
```

```python
def plot_n_images(X,n):
    # 输入的n开方必须为整数
    grid_size = int(np.sqrt(n))
    
    pic_size = int(np.sqrt(X.shape[1]))
    
    first_n_img = X[:n,:]
    
    fig,ax = plt.subplots(nrows=grid_size,ncols=grid_size,sharex=True,sharey=True,figsize=(8,8))
    for i in range(grid_size):
        for j in range(grid_size):
            ax[i,j].imshow(first_n_img[i*grid_size+j,:].reshape((pic_size,pic_size)))
            plt.xticks(np.array([]))
            plt.yticks(np.array([]))
```

```python
plot_n_images(X,9)
plt.show()
```
