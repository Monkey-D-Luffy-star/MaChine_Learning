```python
# 下一步为每一个特征拟合一个高斯分布，画出等高线，分类
def estimate_gaussian(X):
    mu = np.sum(X,axis=0)/X.shape[0]
    sigma = np.sum((X-mu)**2,axis=0)/X.shape[0]
    return mu,sigma
```

```python
mu,sigma = estimate_gaussian(X)
mu,sigma
```

```python
def gaussian(sigma, x, u):
    y = np.exp(-(x - u) ** 2 / (2 * sigma ** 2)) / (sigma * np.sqrt(2 * np.pi))
    return y
```

```python
# 进行数据可视化
# 创造一个网格数据，计算每个点的概率密度，如果在某个阈值，就画一条等高线
xplot = np.linspace(0,25,100)
yplot = np.linspace(0,25,100)
Xplot,Yplot = np.meshgrid(xplot,yplot)
Z = gaussian(sigma[0],Xplot,mu[0])*gaussian(sigma[1],Yplot,mu[1])
Z
```

```python
fig, ax = plt.subplots(figsize=(12,8))
contour = plt.contour(Xplot, Yplot, Z,[7*10**-9, 6*10**-7, 5*10**-5, 1.2*10**-3, 0.016],colors='k')
ax.scatter(X[:,0], X[:,1])
ax.clabel(contour, inline=True, fontsize=10)
plt.show()
```

```python
# 通过交叉验证集的数据来选择epsilon
Xval.shape,yval.shape
```

```python
from scipy import stats
dist = stats.norm(mu[0],sigma[0])
dist.pdf(15)
```

```python
dist.pdf(Xval[:,0])[:50]
```

```python
p = np.zeros((X.shape[0],X.shape[1]))
p[:,0] = stats.norm(mu[0],sigma[0]).pdf(X[:,0])
p[:,1] = stats.norm(mu[1],sigma[1]).pdf(X[:,1])
p = (p[:,0]*p[:,1]).reshape(p.shape[0],1)
```

```python
pval = np.zeros((Xval.shape[0],Xval.shape[1]))
pval[:,0] = stats.norm(mu[0],sigma[0]).pdf(Xval[:,0])
pval[:,1] = stats.norm(mu[1],sigma[1]).pdf(Xval[:,1])
pval.shape
```

```python
# 写一个能选出最佳epsilon的函数
def select_threshold(pval,yval):
    best_epsilon = 0
    best_f1 = 0
    f1 = 0
    
    step = (pval.max()-pval.min())/1000.
    
    for epsilon in np.arange(pval.min(),pval.max(),step):
        preds = pval<epsilon
        
        tp = np.sum(np.logical_and(preds == 1, yval == 1)).astype(float)
        fp = np.sum(np.logical_and(preds == 1, yval == 0)).astype(float)
        fn = np.sum(np.logical_and(preds == 0, yval == 1)).astype(float)
        
        precision = tp/(tp+fp)
        recall = tp/(tp+fn)
        f1 = 2*precision*recall/(precision+recall)
        if f1>best_f1:
            best_f1 = f1
            best_epsilon = epsilon
            pass
        pass
    return best_f1,best_epsilon
        
```

```python
pval = (pval[:,0]*pval[:,1]).reshape(pval.shape[0],1)
best_f1,best_epsilon = select_threshold(pval,yval)
best_f1,best_epsilon
```

```python
outliers = np.where(p<best_epsilon)
outliers
```

```python
fig,ax = plt.subplots(figsize=(12,8))
ax.scatter(X[:,0], X[:,1])
ax.scatter(X[outliers[0],0], X[outliers[0],1],color='r')
plt.show()
```

```python
# 序列化矩阵
def serialize(X,theta):
    return np.concatenate((X.ravel(),theta.ravel()))

def deserialize(param,nusers,nmovies,nfeatures):
    return param[:nmovies*nfeatures].reshape((nmovies,nfeatures)),param[nmovies*nfeatures:].reshape((nusers,nfeatures))


def cost(params,Y,n_features,R):
    X,theta = deserialize(params,Y.shape[1],Y.shape[0],n_features)
    
    inner = np.power((X.dot(theta.T)-Y),2)
    return np.sum(inner[R==1])/2
 
```

```python
path = 'D:/Google_Download/andrew_ml_ex85033/ex8_movieParams.mat'
params_data = loadmat(path)
X = params_data['X']
theta = params_data['Theta']
X.shape,theta.shape
```

```python
users = 4
movies = 5
features = 3
X_sub = X[:movies,:features]
theta_sub = theta[:users,:features]
Y_sub = Y[:movies,:users]
R_sub = R[:movies,:users]
params_sub = serialize(X_sub,theta_sub)
cost(params_sub,Y_sub,features,R_sub)
```

```python
params = serialize(X,theta)
cost(serialize(X,theta),Y,X.shape[1],R)
```

```python
def gradient(params,Y,n_features,R):
    X,theta = deserialize(params,Y.shape[1],Y.shape[0],n_features)
    
    # 这里需要注意，X@theta.T之后是一个和Y同型的矩阵，但是要把未评价的地方置0，就需要乘R,不置0会影响下面矩阵相乘的结果
    inner = np.multiply(X@theta.T,R)-Y
    X_grad = inner@theta
    theta_grad = inner.T@X
    
    return serialize(X_grad,theta_grad)
```

```python
X_grad,theta_grad = deserialize(gradient(params,Y,10,R),Y.shape[1],Y.shape[0],10)
X_grad,theta_grad
```

```python
# 注意壹不是L    o(╥﹏╥)o
def regularized_cost(param,Y,n_features,R,l=1):
    reg_term = np.power(param, 2).sum()*(l / 2)

    return cost(param,Y,n_features,R)+reg_term

def regularized_grad(param,Y,n_features,R,l=1):
    return gradient(param,Y,n_features,R)+l*param
```

```python
params = regularized_grad(params,Y,10,R,l=1)
X_grad,theta_grad = deserialize(params,Y.shape[1],Y.shape[0],10)
X_grad,theta_grad
```
